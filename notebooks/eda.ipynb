```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Time distribution

```python
sns.kdeplot(
    data=df,
    x='Time',
    hue='Class',
    common_norm=False,
    fill=True,
    alpha=0.5,
    palette={1: 'blue', 0: 'pink'}
)
plt.title("Transacciones en el tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Num Transacciones")
plt.legend(title='Class', labels=['Fraud', 'Not Fraud'])
plt.show()
```

Podemos observar que en los que **no son fraude**, la grafica tiene dos picos claros (durante el dia). Sin embargo lo que **si es fraude** es bastante mas lineal en el tiempo, lo que indica que los ciberdelincuentes actuan tambien durante la noche.

### Feature Correlation

```python
corr = df.corr(numeric_only=True)
sns.heatmap(
    data=corr,
    cmap='Oranges',
    linewidths=0.5,
    xticklabels=corr.columns,
    yticklabels=corr.columns
)
```

Entre features V1-V28 no hay ninguna correlacion directa, sin embargo si que podemos apreciar algo de correlacion entre alguna V-Feature y Amount.

Las que mas correlacion y las que vamos a graficar son V20 y V7 con Amount.

```python
sns.lmplot(
    data=df,
    x='V20',
    y='Amount',
    hue='Class',
    fit_reg=True,
    scatter_kws={'s':2}
)
```

```python
sns.lmplot(
    data=df,
    x='V7',
    y='Amount',
    hue='Class',
    scatter_kws={'s':2}
)
```

Se observa que las variables V7 y V20 presentan una correlaciÃ³n positiva con Amount, lo que significa que, a medida que aumentan sus valores, el monto de la transacciÃ³n tambiÃ©n tiende a aumentar.
AdemÃ¡s, las transacciones no fraudulentas muestran un incremento mÃ¡s marcado, mientras que en las fraudulentas el aumento es mucho mÃ¡s leve.

```python
sns.lmplot(
    data=df,
    x='V2',
    y='Amount',
    hue='Class',
    scatter_kws={'s':2}
)
```

```python
sns.lmplot(
    data=df,
    x='V5',
    y='Amount',
    hue='Class',
    scatter_kws={'s':2}
)
```

Las variables V2 y V5 muestran una correlaciÃ³n negativa con Amount, lo que significa que, cuando sus valores aumentan, el monto de la transacciÃ³n tiende a disminuir.
En otras palabras, las transacciones con montos mÃ¡s altos suelen tener valores mÃ¡s bajos en estas variables.
AdemÃ¡s, las lÃ­neas de tendencia para las transacciones fraudulentas son casi planas, indicando que esta relaciÃ³n es mucho menos evidente en los fraudes.

Esto indica que el Amount no es una variable aislada: parte de su variabilidad estÃ¡ contenida en estas componentes.
Por tanto, el modelo puede aprovechar esa relaciÃ³n para detectar patrones de comportamiento financiero que distingan operaciones legÃ­timas de fraudulentas.

Aun asÃ­, las correlaciones con Class son bajas, por lo que la separaciÃ³n no va a venir de una sola variable sino de la combinaciÃ³n de varias.

### Capacidad de distincion entre clases segun variable

```python
var = df.columns.values

i = 0
t0 = df.loc[df['Class'] == 0]
t1 = df.loc[df['Class'] == 1]

plt.figure(figsize=(12,12))
for feature in var:
    i += 1
    plt.subplot(8,4,i)
    sns.kdeplot(t0[feature], bw_adjust=0.5, label='Class 0')
    sns.kdeplot(t1[feature], bw_adjust=0.5, label='Class 1')
    plt.xlabel(feature)
    locs, labels = plt.xticks()
    plt.tick_params(axis='both', which='major', labelsize=8)

plt.tight_layout()
plt.show()
```

Esta grÃ¡fica muestra, para cada variable, la distribuciÃ³n de las transacciones fraudulentas y no fraudulentas.
Cuando ambas curvas presentan una forma similar o se superponen, significa que esa variable no distingue bien entre fraude y no fraude.
Por el contrario, las variables cuyas distribuciones difieren notablemente entre clases son las que mÃ¡s pueden ayudar al modelo a identificar patrones de fraude.
